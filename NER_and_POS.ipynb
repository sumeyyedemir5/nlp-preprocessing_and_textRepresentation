## Named Entity Recognition (NER)

```python
import pandas as pd
import spacy
nlp = spacy.load("en_core_web_sm")
content = "John works at Microsoft and lives in New York. He visited the National History Museum."
doc = nlp(content)
for ent in doc.ents:
  print(ent.text, ent.start_char, ent.end_char, ent.label_)
entities = [(ent.text, ent.label_,ent.lemma_) for ent in doc.ents]
df = pd.DataFrame(entities, columns=["Text", "type","lemma"])
```

```python
word = "Ankara is great city in Turkey"
doc = nlp(word)
for token in doc:
  print(token.text)
  print(token.lemma_)
  print(token.pos_)
  print(token.tag_)
  print(token.dep_)
  print(token.shape_)
  print(token.is_alpha)
  print(token.is_stop)
  print(token.morph)
  print(" ")
  '''
  books
  book
  NOUN
  NNS
  ROOT
  xxxx
  True
  False
  Number=Plur
  '''
```

## Part of Speech (POS) tagging

Metindeki her bir kelimenin dilbilgisel kategorisini belirleyerek cümlenin yapısını analiz etmeyi sağlar.

```python
import spacy
nlp = spacy.load("en_core_web_sm")
content = "What is the weather like today"
doc = nlp(content)
for token in doc:
  print(token.text, token.pos_)
```

```python
content = "I went to the store, but they were closed, so I had to go to another store"
doc = nlp(content)
for token in doc:
  print(token.text, token.pos_)
```
